```python
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions

# Load and preprocess the image
image1 = tf.keras.preprocessing.image.load_img('pic1.jpeg', target_size=(224, 224))
image1 = tf.keras.preprocessing.image.img_to_array(image1)
image1 = tf.keras.applications.vgg16.preprocess_input(image1)  # Use VGG16's preprocess_input function

# Expand dimensions and predict
image1 = tf.expand_dims(image1, axis=0)
yhat1 = model.predict(image1)

# Decode predictions
predictions1 = decode_predictions(yhat1, top=10)

# Display predictions
for pred in predictions1[0]:
    class_label, class_description, probability = pred
    print(f"Class: {class_label}, Description: {class_description}, Probability: {probability}")
```
